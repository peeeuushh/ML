```python
d = data['aggregations']['time']['buckets']
d
```

```python
new_data=[]

for i in d:
    for j in i['process']['buckets']:
        new_data.append({'key_as_string' : i['key_as_string'],
                         'key': j['key'],
                         'doc_count': j['doc_count'],
                         'utilization': j['utilization']['value']
                         })
        
new_data
```

```python
df=pd.DataFrame(new_data)
```
